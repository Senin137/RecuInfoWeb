```python
df_corr = df.corr()
```

```python
sns.heatmap(df_corr,annot=True,fmt='0.2f',linewidths=1)
plt.show()
```

```python
sns.countplot(x='Outcome',data=df,palette=['green','gold'])
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1,3,1)
plt.title('Counter plot')
sns.countplot(x='Pregnancies',data = df,palette=sns.color_palette("Set2"))

plt.subplot(1,3,2)
plt.title('Distribution plot')
sns.distplot(df['Pregnancies'],color='red')

plt.subplot(1,3,3)
plt.title('Box plot')
sns.boxplot(y = df['Pregnancies'])
```
